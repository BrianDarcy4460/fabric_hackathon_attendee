# Learning Lab: Student Success Data Agent
This lab follows **6 - Student Success Data Agent** without supplying SDK or API code. Complete
the prompts in the Fabric portal and document your decisions in new markdown cells.
**Prerequisites:** `StudentDemo_Ontology` is available in the workspace, its bindings have
finished processing, the workspace has Copilot capacity, and Data Agent tenant settings are on.

## Prompt 1: Define the agent's job
Create a data agent named for student success. State its purpose in one sentence, identify its
intended users, and define the questions it should answer about retention, GPA, degree progress,
academic engagement, and graduate-exam outcomes.
List topics the agent must decline, such as billing, unsupported personnel judgments, or data
outside this academic dataset.
**Ask Copilot:** `Help me define a Student Success data agent. Do not give me SDK code. Draft a
concise purpose statement, intended users, supported question categories, and appropriate
out-of-scope topics for an academic student-success dataset.`

## Prompt 2: Ground the agent in the ontology
Attach `StudentDemo_Ontology` as the data source. Inspect its student, program, section, course,
professor, and term entities and the relationships between them.
Verify the ontology can traverse from a student to program, enrolled sections, courses, terms,
and professors. Do not publish until the binding is ready.
**Ask Copilot:** `I am grounding a Fabric data agent in StudentDemo_Ontology. Without code,
help me identify the entities, properties, and relationships I should verify before publishing
an agent that answers student-success questions.`

## Prompt 3: Write business instructions
Write instructions that define:
- Retained or persisting students.
- At-risk students.
- Students on track to graduate.
- Graduate-exam success.
Name the measures the agent should use: cumulative GPA, credits earned, percent to degree,
enrollment status, exam score, passing threshold, and exam pass/fail. Explain how the agent
should handle an ambiguous request and when it should ask for clarification.
**Ask Copilot:** `Help me draft natural-language instructions for a Student Success data agent.
Do not provide code. Define retained, at risk, on track to graduate, and graduate-exam success
using GPA, credits, percent to degree, status, and exam data. Include guidance for ambiguous
requests and clarifying questions.`

## Prompt 4: Add data-source guidance
Add concise ontology guidance that tells the agent which entity and relationship to use for
student-performance questions, program comparisons, course and section questions, instructor
questions, and term-based trends.
Include a warning not to interpret salary or other sensitive attributes as a performance
judgment.
**Ask Copilot:** `Help me write concise data-source instructions for a student-success ontology.
Do not write code. Explain how the agent should traverse from students to programs, sections,
courses, terms, and professors, and identify sensitive attributes it must not use for
performance judgments.`

## Prompt 5: Publish and evaluate
Publish the agent, then test it with:
1. Which programs have the largest group of students at risk of not graduating?
2. Compare retention signals for withdrawn and retained students.
3. Which sections have the highest cost per completed credit?
4. Which instructors have the strongest review sentiment?
5. Which students are close to graduation but have not passed their graduate exam?
For each answer, inspect the grounding or generated query, compare it to the corresponding Gold
business mart, and refine instructions where the interpretation is incorrect or incomplete.
**Ask Copilot:** `Help me evaluate this Student Success data-agent response: [paste the user
question, answer, and grounding/query]. Do not rewrite the answer for me. Identify possible
grounding errors, missing assumptions, validation checks against Gold marts, and instruction
changes that would improve the agent.`

## Reflection
Explain how the Gold MLVs and ontology work together: MLVs provide governed, refreshable
business metrics, while the ontology provides the vocabulary and relationships needed for
natural-language questions.
