# Learning Lab: Gold Business Marts with Materialized Lake Views
This lab follows **5 - Gold - Business Marts (MLV)** without providing implementation code.
Complete each prompt in a new SQL or PySpark cell. Attach the notebook to `GoldLakehouse` and
confirm the `silver` schema contains the OneLake shortcuts from `SilverLakehouse`.

## Prompt 1: Design the MLV dependency graph
Inspect the `silver` tables. Identify their grain, primary keys, and join keys. Draw a
dependency graph with four Tier 1 marts: student-term fact, section economics, enriched course
reviews, and student profile. Add Tier 2 marts for retention, graduation progress, section cost,
sentiment by grade, instructor sentiment, and graduate-exam outcomes.
Explain why Tier 2 marts should aggregate Tier 1 marts instead of repeating expensive joins.
**Ask Copilot:** `Inspect the tables in the silver schema and help me design a Materialized Lake
View dependency graph. Do not write code. For each proposed Tier 1 and Tier 2 mart, explain its
grain, source tables, join keys, dependencies, and the business question it answers.`

## Prompt 2: Build the student-term foundation mart
Create a Materialized Lake View at one row per student per term. Combine student-term status,
student, and program data. Include GPA, credits, attendance, LMS activity, advising visits,
degree requirements, and student demographics.
Derive percent-to-degree and a rank that identifies each student's latest term. Add a business
description, then validate uniqueness at the student-and-term grain.
**Ask Copilot:** `Help me plan a Materialized Lake View named student_term_fact at one row per
student per term. Do not give me SQL. Tell me which silver tables and columns I need, which
derived metrics I should define, how to identify a student's latest term, and how to validate
the grain after I write the query.`

## Prompt 3: Build section economics
Create a Materialized Lake View at one row per course section. Combine course section, course,
term, professor, and enrollment data. Derive total section cost, completed credits, and cost per
completed credit. Decide and document how your calculation handles no completed credits.
Validate it by finding the highest and lowest cost-per-credit sections.
**Ask Copilot:** `I am creating a section_economics Materialized Lake View. Without writing
code, explain the joins, aggregation grain, definition of completed credits, calculation for
cost per completed credit, null or divide-by-zero decision, and validation checks I should use.`

## Prompt 4: Build enriched course reviews
Create a Materialized Lake View at one row per AI-enriched course review. Preserve all reviews
even when related enrollment or instructor data is absent. Add grade band, instructor,
department, and course attributes.
Add a data-quality rule for invalid sentiment values. Compare source and mart counts, then
inspect the sentiment distribution by grade band.
**Ask Copilot:** `Guide me through designing a review_enriched Materialized Lake View without
code. Explain how to preserve unmatched reviews, derive grade bands, enrich with instructor and
department attributes, enforce valid sentiment values, and validate the completed mart.`

## Prompt 5: Build the student profile
Create a Materialized Lake View that joins each student to their latest academic progress and
graduate-exam result. Include program, cumulative GPA, cumulative credits, percent to degree,
exam score, passing threshold, pass/fail, and relevant GPA.
Explain why this mart should depend on the student-term foundation mart.
**Ask Copilot:** `Help me design a student_profile Materialized Lake View. Do not provide code.
Explain which fields should come from student, program, latest student-term progress, and
graduate-exam data, and explain why it should reuse the student_term_fact mart.`

## Prompt 6: Build business-answer marts
Create a narrow Tier 2 Materialized Lake View for each question:
1. Which first-term signals distinguish withdrawn students from retained students?
2. How does credit progress differ by current student status?
3. Which sections have the highest cost per completed credit?
4. How does review sentiment vary by grade band?
5. Which instructors have the strongest and weakest review sentiment?
6. How do undergraduate progress and GPA relate to graduate-exam performance?
For every mart, document its grain, metric definitions, intended stakeholder, and a validation
query.
**Ask Copilot:** `For the student-success questions below, help me turn each into a narrow Tier
2 Materialized Lake View. Do not write SQL. For each one, identify the Tier 1 input, grain,
dimensions, measures, filters, stakeholder, and validation approach: [paste one question].`
